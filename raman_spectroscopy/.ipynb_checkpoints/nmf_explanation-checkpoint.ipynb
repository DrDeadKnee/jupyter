# Detailed Investigation of Matrix Decomposition methods in Solute-Correlated Raman Spectroscopy

Recently, the non-negative matrix decomposition method for two components as outlined by [Lawton & Sylvestre (Technometrics, 1971)][1] has seen a great deal of use as a method of determining the perturbations induced by addition of a solute to a liquid. Our lab has conducted similar experiments to those described [elsewhwere][2],
to a mixture of results. This notebook outlines a detailed investigation of perturbations to pure water spectra induced by phenylethyl alcohol using similar methods. However, rather than simply accepting the default output of Lawton & Sylvestre's algorithm, we seek to examine the validity of this method for situations where it is not clear that a single solution can be found.

In essence, we will be taking a dataset composed of a miuxture of water and hydrocarbon spectra, which looks like this:
![Image of several raman spectra, each with a different ratio of hydrocarbon to water](raw_data.png "raw data")

and finding two factors which look like this:
![Image of two seperate vectors which look like independent raman spectra](two_factors.png "two factors")

Before beginning, we will go over some of the fundamentals underpinning the algorithm, and detail exactly what we hope to gain by playing with these assumptions. In order, the algorithm executes the following:
- Construction and decomposition of a covariance matrix (essentially PCA)
- Determining regions of nonnegativity in space of principle components
- Finding the intersections of nonnegative boundaries with normalization criteria line
- Choosing a specific solution from the intersections

This last point is particularly relevant to this analysis, as it is not clear, to this author at least, that the criteria put forth in [the original paper][1] are always appropriate to apply [as is currently being done][2].

[1]: http://amstat.tandfonline.com/doi/abs/10.1080/00401706.1971.10488823 "Lawton & Sylvestre"
[2]: https://scholar.google.com/citations?hl=en&user=CsBrI34AAAAJ&view_op=list_works&sortby=pubdate "Ben Amotz Scholar search"

### Construction and decomposition of a covariance matrix
The above is a six word description of Principal Component Analysis (PCA). In brief, PCA consists of constructing a covariance matrix $\mathbf{C}$ using a set of vectors with identical lengths. The highest ranked eigenvectors of the covariance matrix then used to describe the entire dataset. This allows PCA to be used as a data visualization / data reduction technique, whereby each vector in the original data is represented by only a small number of scalars, corresponding to the amount of each eigenvector is contained with the sample.

The technique is useful because of the following properties of square matrices:

- an $N \times N$ square matrix has $N$ real eigenvectors
- these eigenvectors are all orthogonal (i.e. $\vec{V_i} \cdot \vec{V_j} \quad \forall i \neq j$)
- when reconstructing the origninal matrix with a linear combination of $i < N$ vectors, you will achieve the lowest possible squared residual error if you use the $i$ highest ranked eigenvectors


The first point means that PCA will always 'work', insofar as the input data is complete. Orthogonality means that each vector is uncorrelated from each other vector, and therefor very likely represents some different aspect of the data. The fact that taking only a few eigenvectors allows a linear recombination with provable minimal square error means that, if noise in the original data is gaussian, the first few entries in the matrix are all that is required.

In this particular case, we are assuming that each row of our data (each raman spectrum) is produced as a linear combination of the two component spectra, which we will call $\vec{A}_1$ and $\vec{A}_2$. We want to know what $\vec{A}_1$ and $\vec{A}_2$ look like, without having access to either independently. 

To do so we need to name all of the variables we are going to use. Beginning with a data matrix $\mathbf{D}$ where each row is a different spectrum (each of $N$ columns corresponding to a different wavenumber), we construct the covariance matrix 

$\mathbf{C} = \mathbf{D}^{\mathrm{T}}\mathbf{D}$

Lets take a look at the data and the covariance matrix.

```python
# waveN, D = dbh.load_data_matrices(cnx, set_id=3, filenumbers=range(4))
raw_waveN, raw_D = dbh.load_data_matrices(cnx, set_id=5)
# waveN, D = dbh.load_data_matrices(cnx, set_id=5, filenumbers=range(1, 5))
mask = raw_waveN[0] > 2500
waveN, D = [i[mask] for i in raw_waveN], [i[mask] for i in raw_D]

D = np.asmatrix(D)
C = np.transpose(D) * D

sns.set_context('paper', font_scale=2)
plt.figure(figsize=(8,6))
plt.title(r'Data Matrix $\mathbf{D}$')
plt.xlabel(r'$\nu$ (cm$^{-1}$)')
plt.ylabel(r'intensity (a.u.)')

for index, row in enumerate(D):
    plt.plot(waveN[index], np.asarray(row)[0])
plt.show()

plt.figure(figsize=(8, 6))
plt.title(r'Covariance matrix $\mathbf{C}$')
sns.heatmap(
    C, xticklabels=False, yticklabels=False, linewidths=0, center=np.amax(D[0])**2 / 2.0
)
plt.show()
```

Now we need to decompose the covariance matrix into $N$ eigenvalues $\lambda$ and $N$ eigenvectors $\vec{V}$

$\mathbf{C}\vec{V}_1 = \lambda_1 \vec{V}_1, \quad \mathbf{C}\vec{V}_2 = \lambda_2 \vec{V}_2,\quad \cdots \quad \mathbf{C}\vec{V}_N = \lambda_N \vec{V}_N$

and we want to use a linear combination of the first two eigenvectors, $\vec{V}_1$ and $\vec{V}_2$ to figure out what the spectra of the two chemical species we added to the solution are. Lets take a look at the eigenvalues and eigenvectors of $\mathbf{C}$ to see what we're dealing with. In particular, lets pay attention to the highest ranked eigenvectors. We need to be reasonably sure that the first to eigenvectors represent most of the actual information in the data for this to work.

```python
eigenvalues, eigenvectors = np.linalg.eigh(C)
eigenvectors = eigenvectors.transpose()
column_names = ['eigenvalue', 'rank', 'eigenvector', 'wavenumber']

eigenframe = pandas.DataFrame(
    columns=column_names
)

for i in range(1, 5):
    vector = np.asarray(eigenvectors[-i])[0]
    rank = [i] * len(vector)
    value = [eigenvalues[-i]] * len(vector)
    nu = waveN[0]
    miniframe = pandas.DataFrame(
        np.asarray([value, rank, vector, nu]).transpose(),
        columns = column_names
    )
    eigenframe = eigenframe.append(miniframe)
    
grid = sns.FacetGrid(data=eigenframe, col='rank', col_wrap=2, hue='eigenvalue', size=5, aspect=1.25)
grid.map(plt.plot, "wavenumber", "eigenvector").add_legend()
plt.show()
```

It's pretty clear that the first and second eigenvectors both contain meaningful information, as expected. It's also clear that the fourth eigenvector contains no such information. The third eigenvector defeinitely has some information, but it's not clear how much. Looking at the eigenvalues, it's clear that the first two vectors are far, far more important to re-creating the covariance matrix than are the last two. Because these measurements represent a combination of two chemical species, we'll stick with the two component model for now. The ultimate goal is to use the two highest ranked vectors shown above (called $\vec{V}_1$ and $\vec{V}_2$) to construct two spectra we think are representative of the actual chemical components used in our original sample (called $\vec{A}_1$ and $\vec{A}_2$). To do this we need to find a set of scalars $T$, such that

$\vec{A}_1 = T_{11}\vec{V}_1 + T_{21}\vec{V}_2$

$\vec{A}_2 = T_{12}\vec{V}_1 + T_{22}\vec{V}_2$

is a reasonable claim to make.

In the more compact matrix notation, we have

$\mathbf{A} = \mathbf{T} \mathbf{V}$

as the equation we wish to solve. It is convenient to represent this in what I will call $T$ space, where the x-axis represents the contribution of the first eigenvector to our final model, while the y-axis represents the contribution of the second eigenvector. We'll see that our data falls within a very small region of this space.

*Note: If you're familiar with PCA that $\mathbf{A}$ from the above equation is the same $\mathbf{A}$ as the standard matrix decompostion $\mathbf{D} \rightarrow \mathbf{U} \mathbf{\Sigma} \mathbf{V}^{\mathrm{T}} \rightarrow \mathbf{U} \mathbf{\Sigma} \mathbf{T}^{-1} \mathbf{T} \mathbf{V}$ which can be written as a two-matrix equation, consisting of the spectra $\mathbf{T}\mathbf{V}^{\mathrm{T}} = \mathbf{A}$ and the concentrations $\mathbf{U} \mathbf{\Sigma} \mathbf{T}^{-1} = \mathbf{L}$ such that $\mathbf{D} = \mathbf{L} \mathbf{A}$. This entire excercise is about determining $\mathbf{A}$.*

```python
V1 = np.asarray(eigenframe['eigenvector'][eigenframe['rank'] == 1.0])
V2 = np.asarray(eigenframe['eigenvector'][eigenframe['rank'] == 2.0])
V1_contributions = np.asarray([np.dot(V1, np.asarray(i)[0]) for i in D])
V2_contributions = np.asarray([np.dot(V2, np.asarray(i)[0]) for i in D])


def plot_contributions(title=False):
    plt.figure(figsize=(8, 6))
    if title:
        plt.title(title)
    plt.xlabel(r'$T_{1x} = \vec{V}_1 \cdot \vec{D}_x$')
    plt.ylabel(r'$T_{2x} = \vec{V}_2 \cdot \vec{D}_x$')
    for i in range(len(V1_contributions)):
        plt.plot(V1_contributions[i], V2_contributions[i], 'o', color=colors[0])

plt.clf()
plot_contributions(title=r'The dataset $\mathbf{D}$, represented in T-space')
plt.show()
```

### Finding nonnegative combinations of $\vec{V}_1$ and $\vec{V}_2$

Now it bears mentioning that the matrix of eigenvectors $\mathbf{V}$ is automatically given by our data set. This means that all we really need to do is find $\mathbf{T}$ - the desired combination of these vectors that gives us a reasonable set of chemical spectra $\mathbf{A}$. We will accomplish this by analyzing what values $\mathbf{T}$ could possibly contain, by eliminating any combinations that make no physical sense.

We know the two following things about the spectra $\vec{A}$:
- The shape of the spectrum will be given by a ratio of $\vec{V}_1$ to $\vec{V}_2$
- We cannot have less of $\vec{V}_2$ in $\vec{S}_1$ than is present in the sample with the most $\vec{V}_2$
- We cannot have more of $\vec{V}_2$ in $\vec{S}_1$ than is present in the sample with the least $\vec{V}_2$
- $\mathbf{A}$ cannot have any negative entries (corresponding to a negative transition probability)


These statements can be used to construct lines $T_{1x}, T_{1y}$ space which pass throught the origin. The first line will pass through the point with the most $T_2$, while the second will pass through the point with the least $T_2$. Both will pass through the origin. These lines can easily be visualized in $T$-space. We will denote these lines by the symbol $L$, and we will distinguish them by the subscript 1 or 2, corresponding to which of the two chmical species this line will be used to solve for. We will also use the subscript $n$ to denote that these lines are the nearest possible solutions to the $T_2 = 0$ condition.

```python
x = np.arange(0, np.amax(V1_contributions) * 1.1, np.amax(V1_contributions) / 100.0)
nearslope_1 = np.amax(V2_contributions) / V1_contributions[np.argmax(V2_contributions)]
nearslope_2 = np.amin(V2_contributions) / V1_contributions[np.argmin(V2_contributions)]
nearline_1 = nearslope_1 * x
nearline_2 = nearslope_2 * x

def plot_near_lines():
    plt.plot(x, nearline_1, '--', color=colors[2], label=r'$V_{1n}$')
    plt.plot(x, nearline_2, '--', color=colors[3], label=r'$V_{2n}$')
    
plot_contributions()
plot_near_lines()
plt.legend(loc='upper left')
plt.show()
```

Each of these 'inner' lines serve as a one-sided boundary for ratios of $T_2$ to $T_2$ which will ultimately be a row in the matrix $\mathbf{T}$. The other side of the boundary is given by the final statement, that of nonnegativity. We can explain this as follows: To begin with, all of the values of $\vec{V}_1$ will have the same sign, which we will assume by default to be positive. This means that the entries $T_{11}$ and $T_{12}$ have to be positive. More importantly, we have that no single wavenumber can have a ratio of $\vec{V}_1$ to $\vec{V}_2$ which gives a negative intensity at any wavenumber, which transltes to the equation

$0 \leq T_{1x}V_{1i} + T_{2x}V_{2i} \quad \forall \quad i$

where $V_{1i}$ denotes the $i$th element of vector $\vec{V}_1$. As some elements of $\vec{V}_2$ are negative, but none of $\vec{V}_1$ are, one of the points is immediateley given by

$-\frac{\vec{V}_{1i^-}}{\vec{V}_{2i^-}} = \frac{T_{2}}{T_{1}}$

while the other is given by

$\frac{\vec{V}_{1i^+}}{\vec{V}_{2i^+}} = \frac{T_{2}}{T_{1}}$

where $i^+$ and $i^-$ are the indices for the 

```python
farslope_1 = min([np.abs(V1[k] / V2[k]) for k in range(len(V2)) if V2[k] < 0])
farslope_2 = -min([np.abs(V1[k] / V2[k]) for k in range(len(V2)) if V2[k] >= 0])
farline_1, farline_2 = x * farslope_1, x * farslope_2

def plot_far_lines():
    plt.plot(x, farline_1, color=colors[2], label=r'$V_{1f}$')
    plt.plot(x, farline_2, color=colors[3], label=r'$V_{2f}$')
    
plot_contributions()
plot_near_lines()
plot_far_lines()
plt.legend(loc='lower left')
plt.show()
```

### Determining intersection of nonnegative regions with normalization criteria
Because of the normalization criteria applied to the original data (i.e. $1 = w \sum_{i=1}^{N} D_{ij} \quad \forall \quad j$), we require of our solutions that the relative amounts of $\vec{V}_1$ and $\vec{V}_2$ sum to one. This is equivalent to stating that in T-space, the solutions will fall on a line we will call the unitary line:

$1 = T_1\sum_{i=1}^{N}V_{1i} + T_2\sum_{i=1}^{N}V_{2i}$

or, denoting the area of the $i$th vector by $a_i$:

$1 = T_1a_1 + T_2a_2$

which for convenience we will rearrange to a form similar to $T_2 = f(a_2, a_1, T_1)$

$T_2 = \frac{1}{a_2}(1 - T_1a_1)$.

The absolute hard limits on solutions are then given by the intersection of this unitary line with the hard limit lines plotted earlier. This is easy to solve, as the orginal equations are all of the form $T_2 = mT_1$, where m is the slope of $T_2$ vs $T_1$. Using the subscript $f$ to denote the intersection with the 'far' line (far from $T_2 = 0$), while the subscript $n$ denotes the intersection with the 'near' line (near to $T_2 = 0$), the points in T-space corresponding to the intersections may be written as

$P_{1f} = \big((a_1 + m_{1f} a_2)^{-1},\quad m_{1f}(a_1 + m_{1f} a_2)^{-1}\big)$

$P_{1n} = \big((a_1 + m_{1n} a_2)^{-1},\quad m_{1n}(a_1 + m_{1n} a_2)^{-1}\big)$

$P_{2n} = \big((a_1 + m_{2n} a_2)^{-1},\quad m_{2n}(a_1 + m_{2n} a_2)^{-1}\big)$

$P_{2f} = \big((a_1 + m_{2f} a_2)^{-1},\quad m_{2f}(a_1 + m_{2f} a_2)^{-1}\big)$

```python
area_1, area_2 = np.sum(V1), np.sum(V2)
unitary_line = (1 / area_2) * (1 - area_1 * x)
intersections = {
    'P1_far': [(area_1 + farslope_1 * area_2)**-1, farslope_1 * (area_1 +farslope_1 * area_2)**-1],
    'P1_near': [(area_1 + nearslope_1 * area_2)**-1, nearslope_1 * (area_1 + nearslope_1 * area_2)**-1],
    'P2_near': [(area_1 + nearslope_2 * area_2)**-1, nearslope_2 * (area_1 + nearslope_2 * area_2)**-1],
    'P2_far': [(area_1 + farslope_2 * area_2)**-1, farslope_2 * (area_1 + farslope_2 * area_2)**-1]
}

def plot_unitary_line():
    plt.plot(x, unitary_line, color=colors[4], label='unitary line')
    
def plot_intersections():
    executed = False
    for key in intersections:
        if not executed:
            plt.plot(intersections[key][0], intersections[key][1], 'o', color=colors[5], label='Intersection')
            executed=True
        else:
            plt.plot(intersections[key][0], intersections[key][1], 'o', color=colors[5])
    
def plot_T_space():
    plot_contributions()
    plot_near_lines()
    plot_far_lines()
    plot_unitary_line()
    plot_intersections()
    yvals = [intersections[key][1] for key in intersections]
    plt.ylim(min(1.1 * min(yvals), -1.1 * min(yvals)), 1.1 * max(yvals))
    
plot_T_space()
plt.legend(loc='lower right')
plt.title('T-space, with all relevant lines and intersections')
plt.show()

plot_T_space()
plt.title('T-space, close-up on representations')
plt.ylim(-0.001, 0.001)
plt.legend(loc='lower left')
plt.show()
```

### Choosing specific solutions as a combination of intersections
We now have four points in T-space, all falling on the same (unitary) line which we can use to construct our solutions. These points in T-space are simply abstract representations of how much of a given eigenvector is present in the true chemical spectra present in the solution. We can retrieve those spectra as rows in the matrix $\mathbf{A}$.

$\vec{A}_{1f} = \vec{V}_1P_{1f} + \vec{V}_2P_{2f}$

$\vec{A}_{1n} = \vec{V}_1P_{1n} + \vec{V}_2P_{2n}$

$\vec{A}_{2n} = \vec{V}_1P_{2n} + \vec{V}_2P_{2n}$

$\vec{A}_{2f} = \vec{V}_1P_{2f} + \vec{V}_2P_{2f}$

```python
solution_spectra = pandas.DataFrame(
    np.asarray([intersections[key][0] * V1 + intersections[key][1] * V2 for key in intersections]).transpose(),
    columns=[key for key in intersections]
)

plt.figure(figsize=(8,6))
plt.xlabel(r'wavenumber (cm$^{-1}$)')
plt.ylabel(r'intensity (a.u.)')
plt.plot(waveN[0], solution_spectra['P1_far'], label='Component 1 Far Limit', color=colors[2])
plt.plot(waveN[0], solution_spectra['P2_far'], label='Component 2 Far Limit', color=colors[3])
plt.legend(loc='upper right')
plt.show()

plt.figure(figsize=(8,6))
plt.xlabel(r'wavenumber (cm$^{-1}$)')
plt.ylabel(r'intensity (a.u.)')
plt.plot(waveN[0], solution_spectra['P1_near'], label='Component 1 Near Limit', color=colors[2])
plt.plot(waveN[0], solution_spectra['P2_near'], label='Component 2 Near Limit', color=colors[3])
plt.legend(loc='upper left')
plt.show()
```

This is exactly what we should expect - the solution with the largest possible space is the clearly the hydrocarbon spectrum (red), while the solution we have narrowed down is the water spectrum (teal). Now we would very much like to take a single solution, rather than say that the solution probably lies somewhere between the near limit and the far limit for each component.

Lawton and Sylvester propose that single solution for each spectum must be of the form

$\vec{A}_1(\beta_1) = \beta_{1} \vec{A}_{1f} + (1 - \beta_1)\vec{A}_{1n} \quad 0 \leq \beta_1 \leq 1$

$\vec{A}_2(\beta_2) = \beta_{2} \vec{A}_{2f} + (1 - \beta_2)\vec{A}_{2n} \quad 0 \leq \beta_2 \leq 1$

Taken altogether, this means that our final solution spectra are functions of only a single variable, $\beta$, given the entire dataset. Our goal is to find a unique value of $\beta$ which we will denote as $\beta*$, which
gives us the unique matrix $\mathbf{A}^*$ which best represents the spectra of chemical species available. 

Lawton and Sylvester suggest that, in the event that if a wavenumber exists where signal is present from the first chemical species but not the second, we need to take the farmost limit - i.e. that $\beta = 1$. This is equivalent to suggesting that the top spectra in the above plot correctly shows the spectra of both chemical species. This is obviously problematic in this case for the water spectrum (teal), which itself is part of the dataset. It is unclear whether or not this is the case for the red spectrum, although the dip in signal at ~3500 wavenumbers is likely an artifact. Taken together, it's pretty clear that we can't just take $\beta = 1$ as the solution to this dataset. 

Does that mean that we can do nothing? I hope not. Lets begin by taking a look at the various outputs for $\beta$, which is confined between 1 and 0. 

```python
def calculate_spectra(beta, far_spec, near_spec):
    return beta * far_spec + (1 - beta) * near_spec

betas = np.arange(0, 1.01, 0.01)
shortbetas = np.arange(0, 1.2, 0.2)

A1_candidates = [calculate_spectra(b, solution_spectra['P1_far'], solution_spectra['P1_near']) for b in shortbetas]
A2_candidates = [calculate_spectra(b, solution_spectra['P2_far'], solution_spectra['P2_near']) for b in shortbetas]

plt.figure(figsize=(8,6))
plt.title(r'$\vec{A}_1(\beta)$')
plt.xlabel(r'wavenumber (cm$^{-1}$)')
plt.ylabel(r'intensity (a.u.)')
for i in range(len(A1_candidates)):
    plt.plot(
        waveN[0], A1_candidates[i] - i * 0.002 + 0.012, 
        color=sns.dark_palette(colors[2], n_colors=len(shortbetas))[i], label=str(shortbetas[i])
    )
plt.legend(loc='upper right')
plt.show()

plt.figure(figsize=(8,6))
plt.title(r'$\vec{A}_2(\beta)$')
plt.xlabel(r'wavenumber (cm$^{-1}$)')
plt.ylabel(r'intensity (a.u.)')
for i in range(len(A2_candidates)):
    plt.plot(
        waveN[0], A2_candidates[i]  - i * 0.002 + 0.012, 
        color=sns.dark_palette(colors[3], n_colors=len(shortbetas))[i], label=str(shortbetas[i])
    )
plt.legend(loc='upper right')
plt.show()
```

The first question we might ask is 'why is this happening', or 'are Lawton and Sylvester wrong?', and the short answer is that 'it's happenning because of noise, and Lawton and Sylvester are not wrong'. Introducing noise to the system invalidates the criteria that there is a wavenumber at which one chemical species privdes signal, but not the other, and vice versa. This is particularly frustrating in the present case because the original spectra look very clean, and because the signal to noise ratio is not well shared between the two components. 

The appropriate choice for $\vec{A}_1$ is likely somewhere very close to $\beta=0$, while the appropriate choice for $\vec{A}_2$ is likely somewhere between 0.8 and 1. Lets see if we can find a metric that places them there.

```python
def entropy(s, eps=10**-7):
    #assert statement here
    return -np.sum((s + eps) * np.log(s + eps))

A1_candidates = [calculate_spectra(b, solution_spectra['P1_far'], solution_spectra['P1_near']) for b in betas]
A2_candidates = [calculate_spectra(b, solution_spectra['P2_far'], solution_spectra['P2_near']) for b in betas]

A1_ent = [entropy(i) for i in A1_candidates]
A2_ent = [entropy(i) for i in A2_candidates]
A1_inf = [np.var(np.asarray(i)) for i in A1_candidates]
A2_inf = [np.var(np.asarray(i)) for i in A2_candidates]

plt.figure(figsize=(8, 6))
plt.xlabel(r'$\beta$')
plt.ylabel(r'$H(\vec{A}(\beta))$')
plt.plot(betas, A1_ent, label=r'$H(\vec{A}_1)$', color=colors[2])
plt.plot(betas, A2_ent, label=r'$H(\vec{A}_2)$', color=colors[3])
plt.legend(loc='upper right')
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(waveN[0], A1_candidates[np.argmax(A1_ent)], color=colors[2], label=r'$A_1(\beta_1*)$')
plt.plot(waveN[0], A2_candidates[np.argmax(A2_ent)], color=colors[3], label=r'$A_2(\beta_2*)$') 
plt.legend(loc='upper right')
plt.xlabel(r'wavenumber (cm$^{-1}$)')
plt.ylabel(r'intensity (a. u.)')
plt.show()
```

Maximum entropy seems like it produces something near what we suspect $A(\beta*)$ will look like, but it's not clear that, for instance $\vec{A}_2(\beta*)$ is properly distinguishable from any other nearby $\vec{A}_2(\beta)$. 

How might we distinguish these two? Well, we are proposing that one important reason the $\beta = 1$ assumption is invalid is due to the presence of noise in the original data. If we add noise to the data before decomposing it, we should then expect to obtain similar results. Lets see what the far limits look like when we add noise to the data.

# Non-Negative Matrix Independence

A different method of dimensionality reduction is independent component analysis, or ICA. ICA works similarly in principle to PCA, but instead of using high-ranked eigenvectors to reconstruct a matrix, ICA uses 'maximally independent components', or the two components that have the greates statisical distance. 

The issue is that the matrix requires whitening before decomposition - a process which effectively removes the mean.
